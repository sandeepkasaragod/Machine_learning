```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
#loading the "preprocessed.csv" breast cancer pre-processed data generated from breast_cancer-1.ipynb notebook
df = np.loadtxt("preprocessed.csv", delimiter=",")
```

```python
#assigning X and y values
X = df[:,1:31]
y = df[:,0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40)
```

```python
#set up the layer, no modification in the layer
model = Sequential()
model.add(Dense(34, input_dim=30, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=150, batch_size=10)
```

```python
accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy[1]*100))
```
