```python
'''
This is an advanced session to deal with data, here we are dealing with more number of complex features of breast cancer.
Hence, before begin to the modeling, it is good to know about the data, i.e. a primilinary analysis is required to understand 
the data.

The data contains 
    i) clinical information
    ii) gene expression values
    iii) mutation information
    
In this example, I am taking only the clinical information having columns starting from 1 to 31

In the later example (breast_cancer-4) I will be dealing with the the clinical + expression + mutation information 

For this example the data has been taken from 
    "https://www.kaggle.com/datasets/raghadalharbi/breast-cancer-gene-expression-profiles-metabric/code"
'''
```

```python
import pandas as pd
```

```python
df = pd.read_csv("METABRIC_RNA_Mutation.csv", sep=",")
```

```python
df =  df.iloc[:,0:31].copy()
```

```python
df
```

```python
# this data has got many colums, we can try showing only the categorical columns

#store all the column names in a variable columns
columns = df.columns

#get all the numberical columns
numerical_columns = df._get_numeric_data().columns

# get the categorical columns
list(set(columns) - set(numerical_columns))
```

```python
# getting the distinct values from the categorical columns 
print ("cancer_type_detailed", df['cancer_type_detailed'].unique())
print ("pr_status", df['pr_status'].unique())
print ("type_of_breast_surgery", df['type_of_breast_surgery'].unique())
print ("3-gene_classifier_subtype", df['3-gene_classifier_subtype'].unique())
print ("her2_status_measured_by_snp6", df['her2_status_measured_by_snp6'].unique())
print ("inferred_menopausal_state", df['inferred_menopausal_state'].unique())
print ("her2_status", df['her2_status'].unique())
print ("primary_tumor_laterality", df['primary_tumor_laterality'].unique())
print ("er_status", df['er_status'].unique())
print ("integrative_cluster", df['integrative_cluster'].unique())
print ("death_from_cancer", df['death_from_cancer'].unique())
print ("pam50_+_claudin-low_subtype", df['pam50_+_claudin-low_subtype'].unique())
print ("cellularity", df['cellularity'].unique())
print ("er_status_measured_by_ihc", df['er_status_measured_by_ihc'].unique())
print ("tumor_other_histologic_subtype", df['tumor_other_histologic_subtype'].unique())
print ("cancer_type", df['cancer_type'].unique())
```

```python
#droping NaN
df = df.dropna()
```

```python
# converting the above categorical columns to integers
df['cancer_type_detailed'] = df['cancer_type_detailed'].map({'Breast Invasive Ductal Carcinoma':1, \
                                                             'Breast Mixed Ductal and Lobular Carcinoma':2, \
                                                             'Breast Invasive Lobular Carcinoma':3, \
                                                             'Breast Invasive Mixed Mucinous Carcinoma':4, \
                                                             'Breast':5, \
                                                             'Metaplastic Breast Cancer':6})

df['pr_status'] = df['pr_status'].map({'Negative':0, 'Positive':1})

df['type_of_breast_surgery'] = df['type_of_breast_surgery'].map({'BREAST CONSERVING': 1, 'MASTECTOMY': 2})

df['3-gene_classifier_subtype'] = df['3-gene_classifier_subtype'].map({'ER-/HER2-':1, \
                                                                       'ER+/HER2- High Prolif':2, \
                                                                      'GAIN':3, \
                                                                      'HER2+':4})

df['her2_status_measured_by_snp6'] = df['her2_status_measured_by_snp6'].map({'NEUTRAL':1, \
                                                                       'LOSS':2, \
                                                                      'ER+/HER2- Low Prolif':3, \
                                                                      'UNDEF':4})

df['inferred_menopausal_state'] = df['inferred_menopausal_state'].map({'Post':1, 'Pre':0})

df['her2_status'] = df['her2_status'].map({'Negative':0, 'Positive':1})

df['primary_tumor_laterality'] = df['primary_tumor_laterality'].map({'Right':0, 'Left':1})

df['er_status'] = df['er_status'].map({'Negative':0, 'Positive':1})

df['integrative_cluster'] = df['integrative_cluster'].map({'4ER+':1, '4ER-':0})

df['death_from_cancer'] = df['death_from_cancer'].map({'Living':1, 'Died of Disease':0, 'Died of Other Causes':2})

df['pam50_+_claudin-low_subtype'] = df['pam50_+_claudin-low_subtype'].map({'claudin-low': 1, \
                                                                           'LumA': 2, \
                                                                           'LumB': 3,\
                                                                           'Her2': 4, \
                                                                           'Normal': 5, \
                                                                           'Basal': 6, \
                                                                           'NC': 7})

df['cellularity'] = df['cellularity'].map({'High':1, \
                                           'Moderate':2, \
                                           'Low':3})


df['er_status_measured_by_ihc'] = df['er_status_measured_by_ihc'].map({'Negative':0, 'Positive':1})

df['tumor_other_histologic_subtype'] = df['tumor_other_histologic_subtype'].map({'Ductal/NST':1, \
                                                                                 'Mixed':2, \
                                                                                 'Lobular':3, \
                                                                                 'Tubular/ cribriform':4, \
                                                                                 'Mucinous':5, \
                                                                                 'Medullary':6, \
                                                                                 'Metaplastic':7 })


df['cancer_type'] = df['cancer_type'].map({'Breast Cancer':1, 'Breast Sarcoma':2})
```
